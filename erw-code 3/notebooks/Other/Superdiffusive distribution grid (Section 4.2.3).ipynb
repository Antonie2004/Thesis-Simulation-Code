```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

def simulate_erw(n, p, q):
    rng = np.random.default_rng()
    X = np.empty(n, dtype=int)
    X[0] = 1 if rng.random() < q else -1
    for k in range(1, n):
        j = rng.integers(0, k)
        X[k] = X[j] if rng.random() < p else -X[j]
    return np.concatenate(([0], np.cumsum(X)))

n         = 5000
n_paths   = 15000
p_values  = [0.80, 0.95, 0.99]
q_values  = [0.25, 0.50, 0.75]
colors    = ['teal', 'goldenrod', 'indianred']

fig, axes = plt.subplots(3, 3, figsize=(12, 11))

for i, p in enumerate(p_values):
    alpha = 2 * p - 1
    scale = n**alpha / gamma(alpha + 1)

    for j, (q, color) in enumerate(zip(q_values, colors)):
        beta = 2 * q - 1

        Z = np.array([
            simulate_erw(n, p, q)[-1] / scale - beta
            for _ in range(n_paths)
        ])

        ax = axes[i, j]
        ax.hist(Z, bins=60, density=True, color=color, alpha=0.85, edgecolor='none')

        ax.set_xlabel(r'$Z_n$', fontsize=11)
        ax.set_ylabel('density', fontsize=11)

        if i == 0:
            ax.set_title(f'$q = {q}$', fontsize=12)
        if j == 0:
            ax.set_ylabel(f'$p = {p}$\n\ndensity', fontsize=11)

fig.suptitle(
    r'Histograms of $Z_n := S_n/(n^{\alpha}\Gamma(\alpha+1)^{-1}) - \beta$'
    '\n'
    r'over 15,000 paths at $n = 5{,}000$',
    fontsize=13
)

plt.tight_layout()
plt.savefig('9PlotsFigureSection4.2.3.pdf', bbox_inches='tight')
plt.show()
```
